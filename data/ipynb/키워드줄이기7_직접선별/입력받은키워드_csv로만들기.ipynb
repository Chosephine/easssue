## [1] 기존 csv, pickle 가져오기

```python
# 1. 가져오기
import pandas as pd
old_kwd_df = pd.read_csv('./kwd/20221115_02_kwd.csv')

import pickle
with open('./kwd/20221115_02_kwd_name_id.pickle','rb') as f:
    old_kwd_pickle = pickle.load(f)
old_kwd_df
```

```python
# 2. set으로 만들기
old_kwd_set = set(old_kwd_pickle)
old_kwd_set
```

## [2] 팀원 입력 가져오기

```python
# 1. 입력받기
team_kwd_input = ''
while True:
    tmp = input('팀원들에게 받은 것 입력하기 / 그만하려면 q입력 :  ')
    if tmp == 'q':
        break
    team_kwd_input += ' ' + tmp + ','
    print('현재 결과 : ' , team_kwd_input)
```

```python
# 2. set으로 바꾸기
team_kwd_input_lst = team_kwd_input.split(',')
team_kwd_input_lst = [kwd.strip() for kwd in team_kwd_input_lst if kwd!='']
team_kwd_input_set = set(team_kwd_input_lst)
team_kwd_input_set
```

## [4] kwd selection 한 것

```python
#  load 해서, 하나로 합치기
import pickle, glob

folder_name_lst = ['under_three_char','others','over_two_words']

selected_kwd_dict = {}
for folder_name in folder_name_lst:
    print('-'*50)
    print(f'folder name is {folder_name}')

    pickle_path_name_lst = glob.glob(f'./kwd_selection/after_kwd_6th_preprocess/{folder_name}/*.pickle')

    for pickle_path_name in pickle_path_name_lst:
        print(f'pickle name is {pickle_path_name}')
        with open(f'{pickle_path_name}', 'rb') as f:
            cur_kwd_dict = pickle.load(f)
            selected_kwd_dict.update(cur_kwd_dict)  # 기존의 것에 update

print(selected_kwd_dict)
selected_kwd_set = set(selected_kwd_dict)
print(len(selected_kwd_set), selected_kwd_set)
```

## [5] 보통 csv 에서 가져오기
+ https://brikorea.com/bbs/board.php?bo_table=rank&wr_id=219

```python
#브랜드 평판에서 가져오기
# import pandas as pd
# brand1 = pd.read_csv('./kwd_selection/브랜드평판1.csv', encoding='cp949')
# brand1_lst = brand1.iloc[:,1].tolist()
# brand1_set = set(brand1_lst)
# print(brand1_set)

# brand2 = pd.read_csv('./kwd_selection/브랜드평판2_개인.csv', encoding='cp949')
# brand2_lst = brand2.iloc[:,1].tolist()

# brand2_name_lst= [group_name.split()[-1] for group_name in brand2_lst ]
# brand2_name_set = set(brand2_name_lst)
# print(brand2_name_set)

# print(len(brand2_name_set), len(brand1_set))

import re
# 한국은행에서 가져오기
economy_kwd_set = set()
with open('../data/kwd_selection/경제용어.txt') as f:
    total = f.readlines()
    for line in total:
        if len(line.split('・')) >10:
            kwd = line.split('・')[0]
            # print(kwd)
            kwd = re.sub('\(.*?\)' , '' , kwd)
            kwd = kwd.strip()
            # print(kwd)
            if len(kwd) > 1:
                economy_kwd_set.add(kwd)
            # print(line.split('・')[0])

economy_kwd_set
```

## [6] : [1],[2],[3],[4], [5] 을 하나로 합쳐서 csv, pickle 저장하기

```python
# 1. 중복키워드 없애기
new_kwd_set = old_kwd_set.union(team_kwd_input_set, google_kwd_input_set, selected_kwd_set)#, economy_kwd_set)#, economy_kwd_set)
total_len = len(old_kwd_set)+len(team_kwd_input_set)+len(google_kwd_input_set) + len(selected_kwd_set) #+len(economy_kwd_set)#+ len(economy_kwd_set)
duplicated_len = total_len - len(new_kwd_set)
print(f' {total_len} 개에서 중복 {duplicated_len}개 제거하여,  {len(old_kwd_set)} 개 에서 -> {len(new_kwd_set)}개 키워드셋 존재  ')
print(f'새로 추가된 키워드 : ')
print(new_kwd_set - old_kwd_set )
```

```python
# 2. 새로운 kwd 들의 id 부여
## [1] 기존 kwd의 최대 id 찾기
max_kwd_id = old_kwd_df.kwd_id.max()
max_kwd_id += 1 # 새로 시작할 id


## [2] 새로 저장할 pickle 과 df 만들기
import pandas as pd

new_kwd_pickle = old_kwd_pickle.copy()
new_kwd_df = pd.DataFrame()
id_lst = []
name_lst = []
for new_kwd in new_kwd_set:
    if new_kwd not in old_kwd_pickle:  # 기존의 것이 아니라면,
        name_lst.append(new_kwd)
        id_lst.append(max_kwd_id)
        new_kwd_pickle[new_kwd] = max_kwd_id
        max_kwd_id += 1
new_kwd_df['kwd_id'] = id_lst
new_kwd_df['kwd_name'] = name_lst

## [3] df 합치기
new_kwd_df = pd.concat([old_kwd_df,new_kwd_df])
new_kwd_df = new_kwd_df.reset_index(drop=True)
new_kwd_df
```

```python
# 3. df랑 pickle 저장하기
import pickle

while True:
    kwd_filedate = input('kwd_filedate (20221109 꼴) 를 입력해주세요 : ')
    kwd_try = input('kwd_try (01, 02, 03, ... 꼴) 를 입력해주세요 : ')
    check = input(f'{kwd_filedate}_{kwd_try} 꼴이 맞으면 y 아니면 n : ')
    if check == 'y':
        break

## [1] pickle 저장하기
with open(f'./kwd/{kwd_filedate}_{kwd_try}_kwd_name_id.pickle','wb') as f:
    pickle.dump(new_kwd_pickle,f)


## [2] csv 저장하기
new_kwd_df.to_csv(f'./kwd/{kwd_filedate}_{kwd_try}_kwd.csv', index=False, header=True)

print('pickle과 csv가 잘 저장되었습니다.')
```
