```python
after6_over_two_words_dict

import pandas as pd
kwd_meta_df = pd.read_csv('./kwd_selection/after_kwd_6th_preprocess/kwd_meta_over_two_words.csv')

start_idx = kwd_meta_df.loc[0,'over_two_words']
end_idx = 99999999999999



sorted_after6_over_two_words_dict = sorted(after6_over_two_words_dict.items(), key=lambda item: item[1], reverse=True)
sorted_after6_over_two_words_dict   # [ ('전 세계', 18109) , ('바이러스 감염', 15549),  ('두 사람', 7633), ...] 꼴
n = len(sorted_after6_over_two_words_dict)

saved_dict = {}
cur_idx = 0
for i in range(start_idx, n):
    cur_idx = i
    kwd, cnt = sorted_after6_over_two_words_dict[i]
    res = 'init'
    while res not in ['ㅇ','ㄴ','ㄱ']:
        print(f'{cur_idx, kwd, cnt}')
        res = input('살리려면 ㅇ, 아니면 ㄴ , 그만하려면 ㄱ을 누르세요 : ')

        if res == 'ㅇ':
            saved_dict[kwd] = cnt
    
    if res == 'ㄱ':
        print('현재까지 진행한 index : ', cur_idx)
        break
    

    if i % 10 == 0:
        print('현재까지 살린 것의 개수는  : ', len(saved_dict))

end_index = cur_idx if res!= 'ㄱ' else cur_idx-1


# 키워드 저장
with open(f'./kwd_selection/after_kwd_6th_preprocess/over_two_words/over_two_words_{start_idx}_{end_index}.pickle', 'wb') as f:
    pickle.dump(saved_dict,f)

# 키워드 메타 데이터 저장
kwd_meta_df.loc[0,'over_two_words'] = end_index + 1
kwd_meta_df.to_csv('./kwd_selection/after_kwd_6th_preprocess/kwd_meta_over_two_words.csv', header=True, index=False)
```
