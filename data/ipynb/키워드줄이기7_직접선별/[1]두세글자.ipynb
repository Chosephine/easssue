```python
# 키워드 줄이기6 결과물 가져오기
import pickle
with open(f'./kwd_selection/after_kwd_6th_preprocess/under_three_char.pickle', 'rb') as f:
    after6_under_3_chars_dict = pickle.load(f)

import pickle
with open(f'./kwd_selection/after_kwd_6th_preprocess/over_two_words.pickle', 'rb') as f:
    after6_over_two_words_dict = pickle.load(f)

import pickle
with open(f'./kwd_selection/after_kwd_6th_preprocess/others.pickle', 'rb') as f:
    after6_others_dict = pickle.load(f)

import pandas as pd
kwd_meta_df = pd.read_csv('./kwd_selection/after_kwd_6th_preprocess/kwd_meta_under_three_char.csv')


end_idx = 99999999999999
```

```python
after6_under_3_chars_dict

sorted_after6_under_3_chars_dict = sorted(after6_under_3_chars_dict.items(), key=lambda item: item[1], reverse=True)
sorted_after6_under_3_chars_dict   # [ ('전세계', 18109) , ('감염', 15549),  ('두사람', 7633), ...] 꼴
n = len(sorted_after6_under_3_chars_dict)


saved_dict = {}
cur_idx = 0
start_idx = kwd_meta_df.loc[0,'under_three_char']
for i in range(start_idx, n):
    cur_idx = i
    kwd, cnt = sorted_after6_under_3_chars_dict[i]
    res = 'init'
    while res not in ['y','n','s']:
        print(f'{cur_idx, kwd, cnt}')
        res = input('살리려면 y, 아니면 n , 그만하려면 s를 누르세요 : ')

        if res == 'y':
            saved_dict[kwd] = cnt
    
    if res == 's':
        print('현재까지 진행한 index : ', cur_idx)
        break
    

    if i % 10 == 0:
        print('현재까지 살린 것의 개수는  : ', len(saved_dict))


end_index = cur_idx if res!= 's' else cur_idx-1


# 키워드 저장
with open(f'./kwd_selection/after_kwd_6th_preprocess/under_three_char/under_three_char_{start_idx}_{end_index}.pickle', 'wb') as f:
    pickle.dump(saved_dict,f)

# 키워드 메타 데이터 저장
kwd_meta_df.loc[0,'under_three_char'] = end_index + 1
kwd_meta_df.to_csv('./kwd_selection/after_kwd_6th_preprocess/kwd_meta_under_three_char.csv', header=True, index=False)
```
