```python
# 키워드 줄이기6 결과물 가져오기
import pickle
with open(f'./kwd_selection/after_kwd_6th_preprocess/under_three_char.pickle', 'rb') as f:
    after6_under_3_chars_dict = pickle.load(f)

import pickle
with open(f'./kwd_selection/after_kwd_6th_preprocess/over_two_words.pickle', 'rb') as f:
    after6_over_two_words_dict = pickle.load(f)

import pickle
with open(f'./kwd_selection/after_kwd_6th_preprocess/others.pickle', 'rb') as f:
    after6_others_dict = pickle.load(f)

import pandas as pd
kwd_meta_df = pd.read_csv('./kwd_selection/after_kwd_6th_preprocess/kwd_meta_others.csv')

```

```python
after6_others_dict

sorted_after6_others_dict = sorted(after6_others_dict.items(), key=lambda item: item[1], reverse=True)
sorted_after6_others_dict   # [ ('전세계사람', 18109) , ('바이러스', 15549),  ('코로나19', 7633), ...] 꼴
n = len(sorted_after6_others_dict)

saved_dict = {}
cur_idx = 0
start_idx = kwd_meta_df.loc[0,'others']
for i in range(start_idx, n):
    cur_idx = i
    kwd, cnt = sorted_after6_others_dict[i]
    res = 'init'
    while res not in ['ㅇ','ㄴ','ㄱ']:
        print(f'{cur_idx, kwd, cnt}')
        res = input('살리려면 ㅇ, 아니면 ㄴ , 그만하려면 ㄱ를 누르세요 : ')

        if res == 'ㅇ':
            saved_dict[kwd] = cnt
    
    if res == 'ㄱ':
        print('현재까지 진행한 index : ', cur_idx)
        break
    

    if i % 10 == 0:
        print('현재까지 살린 것의 개수는  : ', len(saved_dict))

end_index = cur_idx if res != 'ㄱ' else cur_idx-1


# 키워드 저장
with open(f'./kwd_selection/after_kwd_6th_preprocess/others/others_{start_idx}_{end_index}.pickle', 'wb') as f:
    pickle.dump(saved_dict,f)

# 키워드 메타 데이터 저장
kwd_meta_df.loc[0,'others'] = end_index + 1
kwd_meta_df.to_csv('./kwd_selection/after_kwd_6th_preprocess/kwd_meta_others.csv', header=True, index=False)
```
